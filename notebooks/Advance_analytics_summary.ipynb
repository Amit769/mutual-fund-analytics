## Insight 1: Historical VaR & CVaR Analysis

- Historical Value at Risk (VaR) at the 95% confidence level identifies the potential maximum daily loss under normal market conditions.
- Conditional Value at Risk (CVaR) measures the average loss when returns fall below the VaR threshold, providing a better estimate of downside risk.
- Funds with the lowest VaR and CVaR values exhibit higher downside risk and may require careful monitoring by risk-averse investors.

## Insight 2: Rolling 90-Day Sharpe Ratio

- The rolling 90-day Sharpe Ratio highlights how the risk-adjusted performance of each fund changes over time.
- Funds maintaining consistently high rolling Sharpe ratios demonstrate stable performance with efficient risk management.
- Significant fluctuations in the rolling Sharpe ratio may indicate periods of increased market volatility or changing fund performance.

## Insight 3: Investor Cohort Analysis

- Investors were grouped according to the year of their first investment transaction.
- Cohort analysis reveals differences in average investment amount, cumulative investment, and preferred mutual fund across investor generations.
- Larger and more active cohorts indicate sustained investor participation and stronger long-term engagement.

## Insight 4: SIP Continuity Analysis

- Investors with six or more SIP transactions were evaluated for investment consistency.
- An average gap greater than 35 days was classified as "At Risk", indicating potential discontinuation of regular investments.
- Maintaining consistent SIP intervals reflects stronger investor discipline and improves long-term wealth creation.

## Insight 5: Portfolio Concentration and Fund Recommendation

- The Herfindahl–Hirschman Index (HHI) was used to measure portfolio concentration across equity funds.
- Higher HHI values indicate greater concentration in fewer holdings, while lower values suggest better diversification.
- A simple recommendation engine was developed to suggest the top three mutual funds based on Sharpe Ratio within the selected investor risk category (Low, Moderate, or High), enabling risk-aware investment decisions.
